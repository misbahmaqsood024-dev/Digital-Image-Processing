```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/dip.jpeg")  # load the image

# convert BGR  RGB (because OpenCV stores colors in reverse order)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("Original Colored Image")
plt.axis("off")
```

```python
R = img.copy()
R[:, :, 1] = 0   # remove green channel
R[:, :, 2] = 0   # remove blue channel

plt.imshow(R)
plt.title("Red Component")
plt.axis("off")
```

```python
G = img.copy()
G[:, :, 0] = 0   # remove red
G[:, :, 2] = 0   # remove blue

plt.imshow(G)
plt.title("Green Component")
plt.axis("off")
```

```python
B = img.copy()
B[:, :, 0] = 0   # remove red
B[:, :, 1] = 0   # remove green

plt.imshow(B)
plt.title("Blue Component")
plt.axis("off")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
```

```python
ret, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(binary, cmap="gray")
plt.title("Binary Image")
plt.axis("off")
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("/content/dip.jpeg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# Convert to binary
ret, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(binary, cmap="gray")
plt.title("Binary Image")
plt.axis("off")
```

```python
# connected components
num_labels, labels = cv2.connectedComponents(binary)
```

```python
total_objects = num_labels - 1
print("Total Number of Objects Detected =", total_objects)
```

```python
# create an empty color image
label_colored = np.zeros((labels.shape[0], labels.shape[1], 3))

for label in range(1, num_labels):  # skip background (0)
    label_colored[labels == label] = np.random.rand(3)  # random colors
```

```python
plt.imshow(label_colored)
plt.title("Connected Component Labeled Image")
plt.axis("off")
```
