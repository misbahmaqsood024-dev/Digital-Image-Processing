```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Read the colored image
image = cv2.imread(r"/content/coins.jpeg")  # Replace with your image filename

# Check if the image was loaded successfully
if image is None:
    print("Error: Could not load the image. Please check the file path and ensure the image exists.")
else:
    print("Image loaded successfully.")
```

```python
# Convert image from BGR (OpenCV default) to RGB for display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Split into individual color components
B, G, R = cv2.split(image)
```

```python
# Create grayscale image
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Create binary image using thresholding
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

```python
# Display all images
plt.figure(figsize=(12, 8))
```

```python
plt.subplot(2, 3, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')
```

```python
plt.subplot(2, 3, 2)
plt.imshow(R, cmap='Reds')
plt.title('Red Component')
plt.axis('off')
```

```python
plt.subplot(2, 3, 3)
plt.imshow(G, cmap='Greens')
plt.title('Green Component')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(B, cmap='Blues')
plt.title('Blue Component')
plt.axis('off')
```

```python
plt.subplot(2, 3, 5)
plt.imshow(gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
```

```python
plt.subplot(2, 3, 6)
plt.imshow(binary, cmap='gray')
plt.title('Binary Image')
plt.axis('off')
```

```python
plt.tight_layout()
plt.show()
```
